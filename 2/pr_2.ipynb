```python
import numpy as np
import cmath
import matplotlib.pyplot as plt
import cv2
from PIL import Image
from scipy.fftpack import idct, dct
```

```python
def dct2(a):
    return dct( dct( a, axis = 0, norm='ortho'), axis=1, norm='ortho')

def idct2(a):
    return idct( idct( a, axis = 0, norm='ortho'), axis=1, norm='ortho')
```

```python
def psnr(C: np.ndarray, CW: np.ndarray):
    C_f = C.astype(np.float32)
    CW_f = CW.astype(np.float32)
    return 10*np.log10(255**2 / (np.mean((C_f-CW_f)**2)))
```

```python
img = np.array(Image.open('goldhill.tif'))
img_w = (img + np.random.randint(0, 50)).astype(np.uint8)
plt.imshow(img_w, cmap='gray')

print(psnr(img, img_w))
print(cv2.PSNR(img, img_w))
```
